# Generating masks using graph based models

### Analysis on the Normalised Cut model

```python
# Normalised Cut model
# compactness , which balances the color-space proximity with image space-proximity

import time

k = 151 # image number

compact_Ncut = 55
threshold = cv2.ADAPTIVE_THRESH_GAUSSIAN_C+cv2.THRESH_OTSU

img_object = Image(X_train[k])
mask = np.squeeze(Y_train[k])
step = [250, 300, 350, 400, 450, 500]
    
IoU_step = []
img_segm = []
img_pred = []

start = time.time()

for i in step:
    
    segm_Ncut = i
    img_object.Ncut_segm(compact_Ncut, segm_Ncut)
    img_segm.append(img_object.Ncut_segmentation)
    pred = img_object.Ncut_segm_threshold_mask(threshold)
    
    x = iou_metric(pred, mask, print_table=False)
    y = iou_metric(np.invert(pred), mask, print_table=False)
    if y>x:
        img_pred.append(np.invert(pred))
    else:
        img_pred.append(pred)
                    
    IoU_step.append(max(x,y))
    
end = time.time()
time = end - start
print('Time: {:10.2f}, Average time: {:10.2f}'.format(time, time/len(step)))
```

```python
m = 2; n = 3; width = 11; height = 6
images = img_segm

plt.figure(figsize=(width, height))
for i in range(len(images)):
    plt.subplot(m,n,i+1)
    plt.imshow(images[i], aspect='auto')
    plt.axis('off')
    plt.title('NCut: {} Segm, {:10.2f} IoU'.format(step[i], IoU_step[i]))
plt.show()
```

```python
m = 1; n = 4; width = 13; height = 3
images = [X_train[k], img_segm[2], img_pred[2], mask]
title = ['Original', 'NCut 350 Segments', 'NCut Mask: IoU 0.54', 'Ground Truth']


plt.figure(figsize=(width, height))
for i in range(len(images)):
    plt.subplot(m,n,i+1)
    plt.imshow(images[i], aspect='auto')
    plt.axis('off')
    plt.title('{}'.format(title[i]))
plt.show()
```

### Analysis on the Felsenszwalb model

```python
# Felsenszwalb model

import time

k = 434 # image number

# scale : float, Free parameter. Higher means larger clusters.
# sigma : float, Width of Gaussian kernel used in preprocessing.
# min_size : int, Minimum component size. Enforced using postprocessing.
sigma=0.8; min_size=20
    
threshold = cv2.ADAPTIVE_THRESH_GAUSSIAN_C+cv2.THRESH_OTSU

img_object = Image(X_train[k])
mask = np.squeeze(Y_train[k])
step = [50, 100, 200, 350, 500, 1000]
    
IoU_step = []
img_segm = []
img_pred = []

start = time.time()

for i in step:
    scale = i
    img_object.felzen_segm(scale, sigma, min_size)
    img_segm.append(img_object.felzen_segmentation)
    pred = img_object.felzen_segm_threshold_mask(threshold)
    
    x = iou_metric(pred, mask, print_table=False)
    y = iou_metric(np.invert(pred), mask, print_table=False)
    if y>x:
        img_pred.append(np.invert(pred))
    else:
        img_pred.append(pred)
                    
    IoU_step.append(max(x,y))    

end = time.time()
time = end - start
print('Time: {:10.2f}, Average time: {:10.2f}'.format(time, time/len(step)))
```

```python
m = 2; n = 3; width = 11; height = 6
images = img_segm

plt.figure(figsize=(width, height))
for i in range(len(images)):
    plt.subplot(m,n,i+1)
    plt.imshow(images[i], aspect='auto')
    plt.axis('off')
    plt.title('F: {} cluster, {:10.2f} IoU'.format(step[i], IoU_step[i]))
plt.show()
```

```python
m = 1; n = 4; width = 13; height = 3
images = [X_train[k], img_segm[0], img_pred[0], mask]
title = ['Original', 'Felsenszwalb 50 clusters', 'Felsenszwalb Mask: IoU 0.96', 'Ground Truth']


plt.figure(figsize=(width, height))
for i in range(len(images)):
    plt.subplot(m,n,i+1)
    plt.imshow(images[i], aspect='auto')
    plt.axis('off')
    plt.title('{}'.format(title[i]))
plt.show()
```

### Analysis on the Random Walker model

```python
# Random Walker model

import time

k = 194 # image number

compact_rw=30    
threshold = cv2.ADAPTIVE_THRESH_GAUSSIAN_C+cv2.THRESH_OTSU

img_object = Image(X_train[k])
mask = np.squeeze(Y_train[k])
step = [100, 250, 500, 750, 1000, 1500]
    
IoU_step = []
img_segm = []
img_pred = []

start = time.time()

for i in step: 
    segm_rw = i
    img_object.rw_segm(compact_rw, segm_rw)
    img_segm.append(img_object.rw_segmentation)
    pred = img_object.rw_segm_threshold_mask(threshold)
    
    x = iou_metric(pred, mask, print_table=False)
    y = iou_metric(np.invert(pred), mask, print_table=False)
    if y>x:
        img_pred.append(np.invert(pred))
    else:
        img_pred.append(pred)
                    
    IoU_step.append(max(x,y))    

end = time.time()
time = end - start
print('Time: {:10.2f}, Average time: {:10.2f}'.format(time, time/len(step)))
```

### Running through all images and calculating IoU and the calculation time

```python
# Normalised Cut model
# compactness , which balances the color-space proximity with image space-proximity

import time

compact_Ncut = 55; segm_Ncut = 350
threshold = cv2.ADAPTIVE_THRESH_GAUSSIAN_C+cv2.THRESH_OTSU
    
IoU_NCut = []
img_segm_NCut = []
img_pred_NCut = []

start = time.time()

for i in range(670):
    img_object = Image(X_train[i])
    mask = np.squeeze(Y_train[i])

    img_object.Ncut_segm(compact_Ncut, segm_Ncut)
    img_segm_NCut.append(img_object.Ncut_segmentation)
    pred = img_object.Ncut_segm_threshold_mask(threshold)
    
    x = iou_metric(pred, mask, print_table=False)
    y = iou_metric(np.invert(pred), mask, print_table=False)
    if y>x:
        img_pred_NCut.append(np.invert(pred))
    else:
        img_pred_NCut.append(pred)
                    
    IoU_NCut.append(max(x,y))
    
end = time.time()
time = end - start
print('Time: {:10.2f}, Average time: {:10.2f}'.format(time, time/670))
```

```python
import time

# scale : float, Free parameter. Higher means larger clusters.
# sigma : float, Width of Gaussian kernel used in preprocessing.
# min_size : int, Minimum component size. Enforced using postprocessing.
scale=50; sigma=0.8; min_size=20    
threshold = cv2.ADAPTIVE_THRESH_GAUSSIAN_C+cv2.THRESH_OTSU
    
IoU_F = []
img_segm_F = []
img_pred_F = []

start = time.time()

for i in range(670):
    img_object = Image(X_train[i])
    mask = np.squeeze(Y_train[i])

    img_object.felzen_segm(scale, sigma, min_size)
    img_segm_F.append(img_object.felzen_segmentation)
    pred = img_object.felzen_segm_threshold_mask(threshold)
    
    x = iou_metric(pred, mask, print_table=False)
    y = iou_metric(np.invert(pred), mask, print_table=False)
    if y>x:
        img_pred_F.append(np.invert(pred))
    else:
        img_pred_F.append(pred)
                    
    IoU_F.append(max(x,y))    

end = time.time()
time = end - start
print('Time: {:10.2f}, Average time: {:10.2f}'.format(time, time/670))
```

```python
# Random Walker model

import time

compact_rw=30; segm_rw=1500    
threshold = cv2.ADAPTIVE_THRESH_GAUSSIAN_C+cv2.THRESH_OTSU
    
IoU_RW = []
img_segm_RW = []
img_pred_RW = []

start = time.time()

for i in range(670): 
    img_object = Image(X_train[i])
    mask = np.squeeze(Y_train[i])
    
    img_object.rw_segm(compact_rw, segm_rw)
    img_segm_RW.append(img_object.rw_segmentation)
    pred = img_object.rw_segm_threshold_mask(threshold)
    
    x = iou_metric(pred, mask, print_table=False)
    y = iou_metric(np.invert(pred), mask, print_table=False)
    if y>x:
        img_pred_RW.append(np.invert(pred))
    else:
        img_pred_RW.append(pred)
                    
    IoU_RW.append(max(x,y))    

end = time.time()
time = end - start
print('Time: {:10.2f}, Average time: {:10.2f}'.format(time, time/670))
```
